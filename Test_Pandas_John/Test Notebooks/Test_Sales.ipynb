Create 2018 Residential Sales CSV file

```
import pandas as pd
data = pd.read_excel('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/organizedData/3_sales/retail_sales_2018.xlsx',
                     skiprows = 2, usecols = [0, 1, 2, 4, 7, 8, 9])

data.rename(columns={data.columns[0]:'year',
                     data.columns[1]:'month',
                     data.columns[2]:'util_id',
                     data.columns[3]:'state',
                     data.columns[4]:'revenue',
                     data.columns[5]:'MWh_sold',
                     data.columns[6]:'num_accounts'},
                     inplace=True)

df = pd.DataFrame(data, columns= ['year',
                                  'month',
                                  'util_id',
                                  'state',
                                  'cust_sector_name',
                                  'num_accounts',
                                  'MWh_sold',
                                  'revenue'])

df['cust_sector_name'] ='Residential'

# Omit extraneous rows that tabulate numbers
df = df[df.util_id != 0]
df = df[df.util_id != 88888]

df.to_csv('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/Test_Pandas_John/Post_Panda_Data_Files/sales_2018.csv',
          index = False)
```

Concatenating 2017 data to 2018 and dropping duplicate util_ID's.

```
import pandas as pd
data_one = pd.read_csv('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/Test_Pandas_John/Post_Panda_Data_Files/utility_2018.csv')
data_two = pd.read_csv('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/Test_Pandas_John/Post_Panda_Data_Files/utility_2017.csv')

print(data_one.shape)
print(data_two.shape)

data_merged = pd.concat([data_one,data_two], axis=0)
data_merged = data_merged.drop_duplicates(subset='util_ID', keep='first')

print(data_merged.shape)

data_merged.to_csv('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/Test_Pandas_John/Post_Panda_Data_Files/utility_master_thru_2017.csv',
          index = False)
```

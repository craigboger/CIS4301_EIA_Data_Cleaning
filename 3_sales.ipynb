```
#2018

#Checklist:
#1) Change column names
#2) Change data types names
#3) Change reading file name
#4) Change column name and typeconverter reference for file reading.
#5) Change columns reading from spreadsheet
#6) 
#
#
#

#Define column names to use in a python list.  Match order to the column order in spreadsheet.
sales2018_columns = ['year',
                     'month',
                     'util_id',
                     'state',
                     'revenue',
                     'MWh_sold',
                     'num_accounts']
#Define datatypes to use for the columns.  Specified in the form of a converter.
sales2018_typeConverter = {'year':int,
                            'month':int,
                            'util_id':str,
                            'state':str,
                            'revenue':float,
                            'MWh_sold':float,
                            'num_accounts':int}
#-------------------------RESIDENTIAL-------------------------
columnNames = sales2018_columns
typeConverter = sales2018_typeConverter
sheetCols = 'A:C,E,H,I,J'
sectorName = 'Residential'
#Read in excel sheet to Pandas Dataframe.
df = pd.read_excel('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/organizedData/3_sales/retail_sales_2018.xlsx', #Directory of spreadsheet to read in.
                            sheet_name = 'Sales Ultimate Cust. -States',  #Define the sheet we are reading data from.
                            skiprows = 2,  #Number of rows to skip to get to column headers.
                            usecols = sheetCols,   #Select which columns you want to read and skip.  A:F -> A to F.   Can list many ranges.
                            names = columnNames,   #Define the names of the columns
                            #index_col = ['year','month','util_id'],  #Define the column to use for indexing the dataframe.  In this case, utility IDs.
                            na_values = [' ','.','0','0.0','0.000','88888'],  #Define flags for null values.
                            converters = typeConverter
                            #nrows = 10,  #Define number of rows to read in
                            )
#Drop rows with util_id now set to NaN
df = df.dropna(subset=['util_id'])
#Drop rows with NaN in the util_id, revenue, MWh_sold, or num_accounts.
df = df.dropna(subset=['revenue','MWh_sold','num_accounts'])
#Convert util_id explicitly to int
df = df.astype({'util_id':int})
#Rearrange columns and add cust_sector_name
df['cust_sector_name'] = sectorName
df = df[['year','month','util_id','state','cust_sector_name','num_accounts','MWh_sold','revenue']]

#Write to final dataframe.
sales2018_residential = df.copy(deep=True)

#-------------------------COMMERCIAL-------------------------
columnNames = sales2018_columns
typeConverter = sales2018_typeConverter
sheetCols = 'A:C,E,K,L,M'
sectorName = 'Commercial'
#Read in excel sheet to Pandas Dataframe.
df = pd.read_excel('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/organizedData/3_sales/retail_sales_2018.xlsx', #Directory of spreadsheet to read in.
                            sheet_name = 'Sales Ultimate Cust. -States',  #Define the sheet we are reading data from.
                            skiprows = 2,  #Number of rows to skip to get to column headers.
                            usecols = sheetCols,   #Select which columns you want to read and skip.  A:F -> A to F.   Can list many ranges.
                            names = columnNames,   #Define the names of the columns
                            #index_col = ['year','month','util_id'],  #Define the column to use for indexing the dataframe.  In this case, utility IDs.
                            na_values = [' ','.','0','0.0','0.000','88888'],  #Define flags for null values.
                            converters = typeConverter
                            #nrows = 10,  #Define number of rows to read in
                            )
#Drop rows with util_id now set to NaN
df = df.dropna(subset=['util_id'])
#Drop rows with NaN in the util_id, revenue, MWh_sold, or num_accounts.
df = df.dropna(subset=['revenue','MWh_sold','num_accounts'])
#Convert util_id explicitly to int
df = df.astype({'util_id':int})
#Rearrange columns and add cust_sector_name
df['cust_sector_name'] = sectorName
df = df[['year','month','util_id','state','cust_sector_name','num_accounts','MWh_sold','revenue']]

#Write to final dataframe.
sales2018_commercial = df.copy(deep=True)

#-------------------------INDUSTRIAL-------------------------
columnNames = sales2018_columns
typeConverter = sales2018_typeConverter
sheetCols = 'A:C,E,N,O,P'
sectorName = 'Industrial'
#Read in excel sheet to Pandas Dataframe.
df = pd.read_excel('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/organizedData/3_sales/retail_sales_2018.xlsx', #Directory of spreadsheet to read in.
                            sheet_name = 'Sales Ultimate Cust. -States',  #Define the sheet we are reading data from.
                            skiprows = 2,  #Number of rows to skip to get to column headers.
                            usecols = sheetCols,   #Select which columns you want to read and skip.  A:F -> A to F.   Can list many ranges.
                            names = columnNames,   #Define the names of the columns
                            #index_col = ['year','month','util_id'],  #Define the column to use for indexing the dataframe.  In this case, utility IDs.
                            na_values = [' ','.','0','0.0','0.000','88888'],  #Define flags for null values.
                            converters = typeConverter
                            #nrows = 10,  #Define number of rows to read in
                            )
#Drop rows with util_id now set to NaN
df = df.dropna(subset=['util_id'])
#Drop rows with NaN in the util_id, revenue, MWh_sold, or num_accounts.
df = df.dropna(subset=['revenue','MWh_sold','num_accounts'])
#Convert util_id explicitly to int
df = df.astype({'util_id':int})
#Rearrange columns and add cust_sector_name
df['cust_sector_name'] = sectorName
df = df[['year','month','util_id','state','cust_sector_name','num_accounts','MWh_sold','revenue']]

#Write to final dataframe.
sales2018_industrial = df.copy(deep=True)


#-------------------------CONCATINATE INTO ONE DATAFRAME-------------------------
sales2018 = pd.concat([sales2018_residential,sales2018_commercial,sales2018_industrial], axis=0)
#Create Heirarchical Index
sales2018.set_index(['year','month','util_id','state','cust_sector_name'], inplace=True)
```

##2018 Territory Table

```
#2018

#Need to read data in for both states and territories

#Checklist:
#1) Change column names
#2) Change data types names
#3) Change reading file name
#4) Change column name and typeconverter reference for file reading.
#5) Change columns reading from spreadsheet
#6) 

#Define column names to use in a python list.  Match order to the column order in spreadsheet.
territory2018_columns = ['year',
                         'util_id',
                         'state',
                         'county']
#Define datatypes to use for the columns.  Specified in the form of a converter.
territory2018_typeConverter = {'year':int,
                               'util_id':int,
                               'state':str,
                               'county':str}
#-------------------------RESIDENTIAL-------------------------
columnNames = territory2018_columns
typeConverter = territory2018_typeConverter
sheetCols = 'A,B,D,E'
#Read in excel sheet for state-county data
df = pd.read_excel('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/organizedData/4_territory/Service_Territory_2018.xlsx', #Directory of spreadsheet to read in.
                            sheet_name = 'Counties_States',  #Define the sheet we are reading data from.
                            skiprows = 2,  #Number of rows to skip to get to column headers.
                            usecols = sheetCols,   #Select which columns you want to read and skip.  A:F -> A to F.   Can list many ranges.
                            names = columnNames,   #Define the names of the columns
                            #index_col = ['year','month','util_id'],  #Define the column to use for indexing the dataframe.  In this case, utility IDs.
                            na_values = '',  #Define flags for null values.
                            converters = typeConverter
                            #nrows = 10,  #Define number of rows to read in
                            )

df2 = pd.read_excel('/content/drive/My Drive/CIS4301_EIA_Data_Cleaning/organizedData/4_territory/Service_Territory_2018.xlsx', #Directory of spreadsheet to read in.
                            sheet_name = 'Counties_Territories',  #Define the sheet we are reading data from.
                            skiprows = 2,  #Number of rows to skip to get to column headers.
                            usecols = sheetCols,   #Select which columns you want to read and skip.  A:F -> A to F.   Can list many ranges.
                            names = columnNames,   #Define the names of the columns
                            #index_col = ['year','month','util_id'],  #Define the column to use for indexing the dataframe.  In this case, utility IDs.
                            na_values = '',  #Define flags for null values.
                            converters = typeConverter
                            #nrows = 10,  #Define number of rows to read in
                            )
#Write to final dataframe.
territory2018 = pd.concat([df, df2], axis=0)
```
